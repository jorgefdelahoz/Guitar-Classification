```python
# Prepare connection
os.environ['PATH'] = f'{os.environ["PATH"]}:{os.getcwd()}/drivers'   
os.environ['PATH']

options = Options()
options.set_headless(headless=True)
driver = webdriver.Firefox(firefox_options=options)

# Links Categories Connection
driver.get('https://www.musicstore.com/en_OE/EUR/Guitars/ST-Models/cat-GITARRE-GITEGSTRAT')
# driver.get('https://www.musicstore.com/en_OE/EUR/Guitars/T-Models/cat-GITARRE-GITEGTELE')
# driver.get('https://www.musicstore.com/en_OE/EUR/Guitars/Single-Cut-Models/cat-GITARRE-GITEGPAULA')
# driver.get('https://www.musicstore.com/en_OE/EUR/Guitars/Double-Cut-Models/cat-GITARRE-GITEGSG')
# driver.get('https://www.musicstore.com/en_OE/EUR/Guitars/Signature-Electric-Guitar-Models/cat-GITARRE-GITEGSIGN')
# driver.get('https://www.musicstore.com/en_OE/EUR/Guitars/Other-models/cat-GITARRE-GITEGANDER')
# driver.get('https://www.musicstore.com/en_OE/EUR/Guitars/Semi-Hollow-Models/cat-GITARRE-GITHAES')
# driver.get('https://www.musicstore.com/en_OE/EUR/Guitars/Hollow-Body-Models/cat-GITARRE-GITHAJAZZ')

# close popup
driver.find_element_by_xpath('//*[@id="footerCookieDirective"]/div/div[3]/div/a').click()

# # click select 90
# driver.find_element_by_xpath('//*[@id="productlist-page"]/div[2]/div/div[2]/div/ul[2]/li[3]/a').click()

url=driver.current_url
count=1
# Depending of category, you will need to use one of them to get link
TOTAL_PAGES=int(driver.find_element_by_xpath('//*[@id="productlist-page"]/div[2]/div/div[2]/ul/li[7]').text)
# TOTAL_PAGES=int(driver.find_element_by_xpath('//*[@id="productlist-page"]/div[2]/div/div[2]/ul').text[-1:])

# driver.find_element_by_xpath('//*[@title="to page 2"]').get_attribute('href')


# generate links
links=[f'{url}/{i}' for i in range(1,TOTAL_PAGES)]
links.insert(0, url)

# Close your connetion (it's really important to avoid collapse your RAM)
driver.close()
```

```python
# Guitar Category Scrap

# Make some lists for extractor
price=[]
model=[]
brand=[]
rating=[]
g_info=[]
url_image=[]

# Add count of your page numbers
count=1

# Iterate into your links
for link in links:
    # Prepare Connection
    os.environ['PATH'] = f'{os.environ["PATH"]}:{os.getcwd()}/drivers'
    os.environ['PATH']


    options = Options()
    options.set_headless(headless=True)
    driver = webdriver.Firefox(firefox_options=options)

    driver.get(link)

    # close popup
    driver.find_element_by_xpath('//*[@id="footerCookieDirective"]/div/div[3]/div/a').click()   
    
    # Add control to count the number of elements extracted, and print the Started Time
    count_p=1
    print(f'PÃ¡gina {count} - Starts: {time.ctime()}')
    time.sleep(1)
    
    # Replicate the actions 
    elm=driver.find_element_by_tag_name('html')
    elm.send_keys(Keys.PAGE_DOWN)
    time.sleep(0.5)
    elm.send_keys(Keys.PAGE_DOWN)
    time.sleep(0.5)
    elm.send_keys(Keys.PAGE_DOWN)
    time.sleep(0.5)
    elm.send_keys(Keys.PAGE_DOWN)
    time.sleep(0.5)
    elm.send_keys(Keys.PAGE_DOWN)
    time.sleep(0.5)
    elm.send_keys(Keys.PAGE_DOWN)
    time.sleep(0.5)
    elm.send_keys(Keys.PAGE_DOWN)
    time.sleep(0.5)
    elm.send_keys(Keys.END)
    time.sleep(0.5)

    
    image_element=driver.find_elements_by_class_name('img-responsive')
    for i in image_element:
        z=i.get_attribute('src')
        if z[-3:]=='jpg':
            url_image.append(z)  
    
    for i in driver.find_elements_by_class_name('tile-product'):        
        g_info.append(i.text)

    for i in driver.find_elements_by_class_name('tile-product'):
        url_guitar.append(i.find_element_by_class_name('js-tracking').get_attribute('href'))
    

    
    for guitar_element in range(0,len(driver.find_elements_by_class_name('tile-product'))):
        time.sleep(1)
        try:
            # close newsletter popup
            driver.find_element_by_xpath('//*[@id="newsletterTeaserModal"]/div/div[1]/span[1]').click()
            time.sleep(1)
            pass
        except:
            pass  

        time.sleep(1)
        
         
        # Acess for each guitar for get more details
        driver.find_elements_by_class_name('tile-product')[guitar_element].click()
          
        time.sleep(5)
        elm=driver.find_element_by_tag_name('html')
        elm.send_keys(Keys.END)

        # Brand guitar
        brand.append(driver.find_element_by_xpath('//*[@id="product-page"]/div[1]/div[3]/div[1]/div/div[1]/h1/span[1]').text)
        time.sleep(0.2)
        # Model Guitar
        model.append(driver.find_element_by_xpath('//*[@id="product-page"]/div[1]/div[3]/div[1]/div/div[1]/h1/span[2]').text)     
        
        time.sleep(0.2)  
        try:
            rating.append(driver.find_element_by_xpath('//*[@id="rating-area"]/div/div/span[2]/meta[1]').get_attribute('content'))
            pass
        except:
            rating.append(0)
            pass
        
        
        
        time.sleep(0.1)
        driver.back()
        time.sleep(0.1)
        count_p += 1

    # Check if your extraction its correct
    print(f'{len(url_image)}\n{len(g_info)}\n{len(url_guitar)}\n{len(brand)}\n{len(model)}\n{len(rating)}')
       
    time.sleep(0.5)
    driver.close()
    count+=1
    print(f'Ends: {time.ctime()}')

print('FINISHED')
```

```python
# Clean extracted info of guitars in "g_info" list
ginfo=[re.split('\n',i) for i in g_info]

for i in ginfo:
    if i[0][0]!='(' and i[0][0]!='I':
        i.pop(0)
for i in ginfo:
    if i[0][0]!='(' and i[0][0]!='I':
        i.pop(0)
for i in ginfo:
    if i[0][0]=='I':
        i.insert(0, None)
```

```python
# Clean format of Rating
def clean_rating(data):
    # Change comma separator with dot
    data=data.apply(lambda raw: re.sub(',','.',raw))
    # Change type to float
    data=data.apply(lambda raw: float(raw) if type(raw)!= float else None)
    # Round 2 decimals
    return data.apply(lambda raw: np.round(raw,decimals=2)) 
```

```python
# Make dataframe (for example "strato_data") adding colum names and save it without index
strato_data = pd.DataFrame(ginfo)
strato_data.drop(strato_data.columns[4:], axis = 1, inplace = True) 

strato_data.columns = ['Reviews','Id','Guitar','Price']
strato_data['Url_Guitar']=url_guitar
strato_data['Brand'] = brand
strato_data['Model'] = model
strato_data['Rating'] = rating
strato_data['Url_Image'] = url_image
strato_data['Category']='Stratocaster'

strato_data.to_csv('./data/raw/stratocaster.csv',index=False)
strato_data
```
